```python
# By Groupby make columns by position job
import pandas as pd

file_path = r'C:\Project\PhythonDataScience\dtsc_modul_2_homework\2017_jun_final - 2017_jun_final.csv'

try:
    df = pd.read_csv(file_path)
    print("File successfully loaded!")

    numeric_columns = df.select_dtypes(include=['number'])

    grouped_data = df.groupby('Role')[numeric_columns.columns].mean()
    print("Grouped data by 'Role'(mean of numeric columns):")

    display(grouped_data)

except FileNotFoundError as e:
    print(f"Error: {e}")
```

```python
# Make new DataFrame, find min and max values in Salalry/mounth column
import pandas as pd
file_path = r'c:\Project\PhythonDataScience\dtsc_modul_2_homework\2017_jun_final - 2017_jun_final.csv'

try:
    df = pd.read_csv(file_path)
    print("File successfully loaded!")

    grouped_data = df.groupby('Role').agg({
        'Salary/month': ['min', 'max']
    })

    grouped_data.columns = ['Min Salary', 'Max Salary']
    print("Grouped data by 'Role' with min and max salary:")
    
    display(grouped_data)

except FileNotFoundError as e:
    print(f"Error: {e}")
    
```

```python
# Make function 'fill_avg_salary' wich will be returns the average monthly salary
import pandas as pd

file_path = r'C:\Project\PhythonDataScience\dtsc_modul_2_homework\2017_jun_final - 2017_jun_final.csv'

def fill_avg_salary(row):
    return row["Salary/month"]

try:
    df = pd.read_csv(file_path)
    print("File successfully loaded!")

    df['avg'] = df.apply(fill_avg_salary, axis=1)
    print("DataFrame with 'avg' column:")

    display(df.head())

except FileNotFoundError as e:
    print(f"Error: {e}")
```

```python
# Analyze price and calculate Max, Min, mean, and medain prices
import pandas as pd 
import matplotlib.pyplot as plt

file_path = r'C:\Project\PhythonDataScience\dtsc_modul_2_homework\bestsellers with categories.csv'

try:
    df = pd.read_csv(file_path)
    print("File successfully loaded!")

    df.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre']

    plt.figure(figsize=(8, 6))
    df['price'].plot(kind='hist', bins=10, color='skyblue', edgecolor='black')
    plt.title('Price distribiution')
    plt.xlabel('Price (in dollars)')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

    max_price = df['price'].max()
    min_price = df['price'].min()
    mean_price = df['price'].mean()
    median_price = df['price'].median()

    print(f"Maximum Price: ${max_price}")
    print(f"Minimum Price: ${min_price}")
    print(f"Avarage (Mean) Price: ${mean_price:.2f}")
    print(f"Medain Price: ${median_price}")

except FileNotFoundError as e:
    print(f"Error: {e}")
```
