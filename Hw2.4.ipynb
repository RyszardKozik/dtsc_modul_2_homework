```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# Sample dataset for demonstration purposes (you can replace this with your actual dataset)
file_path = r'C:\Project\PhythonDataScience\dtsc_modul_2_homework\bestsellers with categories.csv'
df = pd.read_csv(file_path)

# Rename columns to snake_case for consistency
df.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre']

# 1. Bar Plot: Average Price by Genre
plt.figure(figsize=(8, 5))
sns.barplot(x='genre', y='price', data=df, estimator='mean', ci=None, palette='Set2')
plt.title('Average Price by Genre', fontsize=16)
plt.xlabel('Genre', fontsize=12)
plt.ylabel('Average Price (USD)', fontsize=12)
plt.grid(True)
plt.show()

# 2. Scatter Plot: Price vs. Reviews
plt.figure(figsize=(8, 5))
plt.scatter(df['reviews'], df['price'], color='coral', alpha=0.6, edgecolor='black')
plt.title('Price vs. Reviews', fontsize=16)
plt.xlabel('Number of Reviews', fontsize=12)
plt.ylabel('Price (USD)', fontsize=12)
plt.grid(True)
plt.show()

# 3. Line Plot: Average Rating by Year
plt.figure(figsize=(8, 5))
avg_rating_by_year = df.groupby('year')['user_rating'].mean()
avg_rating_by_year.plot(kind='line', marker='o', color='blue', linewidth=2)
plt.title('Average Rating by Year', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Rating', fontsize=12)
plt.grid(True)
plt.show()

# 4. Histogram: Distribution of Prices
plt.figure(figsize=(8, 5))
df['price'].plot(kind='hist', bins=10, color='lightgreen', edgecolor='black')
plt.title('Price Distribution', fontsize=16)
plt.xlabel('Price (USD)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True)
plt.show()

# 5. Box Plot: Boxplot of User Ratings by Genre
plt.figure(figsize=(8, 5))
sns.boxplot(x='genre', y='user_rating', data=df, palette='coolwarm')
plt.title('Boxplot of User Ratings by Genre', fontsize=16)
plt.xlabel('Genre', fontsize=12)
plt.ylabel('User Rating', fontsize=12)
plt.grid(True)
plt.show()
```
